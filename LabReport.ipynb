```python
data1 = {
    "ID"  :[1,2,3,4,5,6,7],
    "name": ["Radha", "Krishna", "Hari", "Shyamsundar", "Hanuman", "Ram", "Sita"],
    "age": [20, 25, 30, 35, 40, 45, 50],
    "gender": ["F", "M", "M", "M", "M", "M", "F"]
}
```

```python
df1=pd.DataFrame(data1)
```
